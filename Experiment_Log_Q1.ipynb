###Question 1
When I first received the dataset, first I thought I won't be needing to augment the dataset and increase the number of samples.

I tried use a simple convolution network of low kernel(3) size and 3 Convolution layers and less dense(64 neurons). It was overfitting a lot. It gave 95% training accuracy but only 40% testing accuracy. I had trained it for 50 epochs. The model was reaching convergence at around 25 epochs.

Then, I tried to increase the dense layer to have more neurons and also tried with 2 dense layers(excluding the dense layer for output), which decreased the training accuracy a little(92%) but increased the testing accuracy to around 50%. I trained it for 50 epochs and convergence was reached at around 30.  

Then I increased kernel size as well; (5, 5, 3) and (7, 5, 3) were tried, greater accuracy was achieved at (5, 5, 3) of 92% training accuracy and 55% of testing accuracy. Trained for 50 epochs, convergence around 30. 

Then, I tried to augment the dataset to increase the number of samples to solve overfitting, but after increasing the samples, accuracy went down. Both training and testing accuracy came down to about 45%, no matter the change in dense layers, dropout, or Convolution layers the maximium I was able to take to accuracy to was 62%. 

Then I tried to change optimizer in all theabove networks, but RMSProp and SGD(the two which were tried only performed worse than adam). 
Even changing the learning rate and momentum had no positive effect. 

I tried a different approach. I tried to fork the model. First I decided to classify the samples into 3 ranges - Capital Alphabet, Small Alphabet and Numbers. Then I used 3 separate models to further classify them. The first model achived good accuracy - above 90% in training and above 70% testing. But 70% testing accuracy is still not good enough for a final model. So, had to drop that idea. 

Then, I tried mixed approach. I augmented the data, added droptout layer and increased dense layer size and for convolution layer used different kernel sizes and a mixture of more neurons and less neurons for each.
Then after testing, I was able to achieve 96% training and 87% testing accuracy with augmentation. The configuration used is :

3 Convolution layers - 32, kernel - 5; 64, kernel - 5; 64, kernel - 3; all relu activation. 

3 Maxpooling layers - 2 pool size

2 Dropout layers - 0.05 and 0.1

2 Dense layers - 128 and 62(Output), relu and softmax activation respectively

1 Flatten layer

Adam optimizer with 0.005 learning rate

Categorical Crossentropy loss function

Trained for 60 epochs, convergence achieved at around 45.
